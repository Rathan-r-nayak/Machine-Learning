```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/train.csv")
df.head()
```

## 1. Categorical data
### barplot

```python
df['Survived'].value_counts().plot(kind='bar')
```

### pie chart

```python
df['Survived'].value_counts().plot(kind='pie', autopct="%.2f")
```

## 2. Numerical Data
### histogram

```python
plt.hist(df['Age'],bins=5)
```

### distplot

```python
sns.distplot(df['Age'])
```

```python
sns.histplot(df['Age'])
```

### Box plot

```python
sns.boxplot(df['Age'])
```
