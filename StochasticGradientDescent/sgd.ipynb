## Custom code

```python
class SGD:
    def __init__(self,lr,epoch):
        self.lr = lr
        self.epoch = epoch
        self.coef_ = None
        self.intercept_ = None
    
    def fit(self,x,y):
        self.coef_ = np.ones(x.shape[1])
        self.intercept_ = 0

        for i in range(self.epoch):
            indices = np.random.permutation(x.shape[0])
            for ind in indices:
                yhat = np.dot(x[ind], self.coef_) + self.intercept_
                self.intercept_ = self.intercept_ - self.lr * (-2 * (y[ind] - yhat))

                self.coef_ = self.coef_ - self.lr * (-2 * (np.dot((y[ind] - yhat),x[ind])))
        print(self.intercept_,self.coef_)

        
    def predict(self,x):
        return np.dot(x, self.coef_) + self.intercept_
```

```python
sgd = SGD(0.01, 100)
sgd.fit(X,y)
```
